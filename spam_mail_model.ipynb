```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("mail_data.csv")
data
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
new_data=data.where((pd.notnull(data)),"")
new_data
```

```python
data.shape
```

```python
data.describe()
```

```python
new_data.loc[(new_data["Category"]=="ham"),"Category"]=0
new_data.loc[(new_data["Category"]=="spam"),"Category"]=1
new_data
```

```python
new_data
```

```python
feature=new_data["Message"]
label=new_data["Category"].astype(int)
```

```python
from sklearn.model_selection import train_test_split
train_x_set,test_x_set,train_y_set,test_y_set=train_test_split(feature,label,test_size=.2,random_state=3)
```

```python
print(feature.shape)
print(train_x_set.shape)
print(test_x_set.shape)
print(train_y_set.shape)
print(test_y_set.shape)
```

```python
train_x_set
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
feature_extraction=TfidfVectorizer(min_df=1, stop_words="english", lowercase=True)
train_x_feature=feature_extraction.fit_transform(train_x_set)
test_x_feature=feature_extraction.transform(test_x_set)
```

```python
print(train_x_feature)
print(test_x_feature)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(train_x_feature,train_y_set)
```

```python
list(train_y_set[:5])
```

```python
model.predict(train_x_feature[:5])
```

```python
from sklearn.metrics import accuracy_score
pre1=model.predict(train_x_feature)
accuracy=accuracy_score(train_y_set,pre1)
accuracy
```

```python
from joblib import dump
dump(model,"Mail_model.joblib")
dump(feature_extraction,"vectorizor.joblib")
```
